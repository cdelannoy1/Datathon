```python


df4 = df['qt_flight_duration']
plt1 = pd.concat([df4, df2], axis=1)


plt.rcParams['agg.path.chunksize'] = 1000000 

flight_durations = df4
item_columns = plt1.columns[1:] 
item_data = df[item_columns]

plt.plot(flight_durations, item_data)

plt.xlabel("Flight Duration")
plt.ylabel("Item Sales")
plt.title("Evolution of Item Sold by Flight Duration")

plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```
